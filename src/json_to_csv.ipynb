# Convert the train data

```python
df=pd.read_json(train_file)
df
```

1.   id是文件序号
2.   annotations是每个句子的属性
3.   data是文件所有句子总和


```python
df.loc[0,'annotations']
```

```python
with open(train_file,'r') as f:
    data = json.loads(f.read(), strict=False)

df_nested_list = pd.json_normalize(data,
                  record_path =['annotations','result'],
                  meta_prefix='document_',
                  meta=['id']
                  )
df_nested_list
```

```python
df_nested_list.isnull().sum()
# type	to_name	from_name 三列缺失大量数据
```

```python
train_csv_file='/content/drive/MyDrive/Colab Notebooks/semEval/legalEval/taskA-RR/data/train.csv'
dev_csv_file = "/content/drive/MyDrive/Colab Notebooks/semEval/legalEval/taskA-RR/data/dev.csv"
```

```python
df_nested_list.to_csv(train_csv_file,index=False)
```
